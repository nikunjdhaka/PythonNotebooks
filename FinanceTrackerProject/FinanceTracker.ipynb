```python
class Transaction:

  def __init__(self):
    self.amount = 0
    self.typ = ""
    self.cat = dict()
    self.savings = 0

  def add_transaction(self, amount, typ, cat):
    if typ.lower() == "income":
      self.typ = "income"
      self.amount = self.amount + amount
    elif typ.lower() == "expense":
      self.typ = "expense"
      self.amount = self.amount - amount
    else:
      print("type of transaction should be income/expense")
    if cat.lower() in self.cat:
      self.cat[cat.lower()] = self.cat[cat.lower()] + self.amount
    else:
      self.cat[cat.lower()] = self.amount
    print("\nTransaction Added successfully\n")

  def total_savings(self):
    print("============TOTAL SAVINGS REPORT=================")
    print(f"\ntotal savings : {self.amount}\n")
    print("=================================================")

  def display_report(self):
    print("category : Amount")
    print("=================FINANCIAL REPORT================")
    for cat_item,cat_amt in self.cat.items():
      print(f"{cat_item} : {cat_amt}")
    print("=================================================")
    print("\n")

```

```python
# from tracker import transaction

print("Welcome to the Finance tracker\n ")
transaction_obj = Transaction() # lets create an object for the class Transaction
while True:
  choice = int(input("\nEnter the choice please:\n1 - Add Transaction\n2 - check Financial report\n3 - check Total savings\n4 - to exit\n"))
  if choice == 1:
    a = int(input("\nEnter the amount of Transaction : "))
    t = input("Enter the type of transaction - Income/Expense : ")
    c = input("Enter the catagory of transaction : ")
    transaction_obj.add_transaction(a, t, c)
  elif choice == 2:
    print("The Financial report for the trasactions done yet :>\n")
    transaction_obj.display_report()
  elif choice == 3:
    transaction_obj.total_savings()
  elif choice == 4:
    break
  else:
    print("\n ******Invalid Choice******\n")
```
