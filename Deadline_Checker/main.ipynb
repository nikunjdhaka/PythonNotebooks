# Deadline checker

```python
# Assignment - 1

# TODO - While creating a task, ask user about the deadline

# Start date - Pick automatically from today's date
# Deadline 

# # to-DO

# 1. Task-1 | Days left  | Green-color
# 2. Task-2 | Days left  | Red-color
# 3. Task-3 | Days left  | Green-color
# 4. Task-4 | Days left  | Red-color
```

```python
import pandas as pd
from datetime import datetime, timedelta
from termcolor import colored

# Get today's date
today = pd.to_datetime("today").normalize()

# Ask number of tasks
n = int(input("How many tasks do you want to create? "))

# Create empty DataFrame
df = pd.DataFrame(columns=["Task", "Start Date", "Deadline (days)", "Deadline Date", "Days Left", "Status"])

# Collect task info
for _ in range(n):
    task_name = input("Enter task name: ")
    deadline_days = int(input("Enter deadline in days: "))
    
    deadline_date = today + pd.Timedelta(days=deadline_days)
    days_left = (deadline_date - today).days
    status = "Green" if days_left > 5 else "Red"
    
    # Add to DataFrame
    df = pd.concat([df, pd.DataFrame([{
        "Task": task_name,
        "Start Date": today,
        "Deadline (days)": deadline_days,
        "Deadline Date": deadline_date,
        "Days Left": days_left,
        "Status": status
    }])], ignore_index=True)

# Display tasks with color
print("\n--- TASK LIST ---")
for idx, row in df.iterrows():
    line = f"{idx + 1}. {row['Task']} | Days left: {row['Days Left']}"
    color = 'green' if row['Status'] == 'Green' else 'red'
    print(colored(line, color))
    

```

```python
df.to_csv("task_list.csv", index=False)
```
