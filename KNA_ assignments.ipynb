```python
""" This Assignment is basically based on the concept of Socail media -
here we will be handling certain features related to Login, signup, storing and viewing posts that have been posted
also we will be creating multpile profile by creating multiple objects """


class SocialMedia():
  def __init__(self):
    self.count = 1
    self.dic_post = dict()
    self.db = dict()
  def login(self,username, password):
    self.username = username
    self.password = password
    if self.username in self.db.keys():
      if self.password == self.db[self.username]:
        print(f"login successful for {self.username}")
      else:
        print("password wrong")
    else:
      print("please signup first")

  def signup(self, username,password):
    self.db[username] = password

  def post(self, p):
    self.dic_post[self.count] = p
    self.count = self.count + 1

  def see_all_posts(self):
    for i, j in self.dic_post.items():
      print(f"post {i} :> {j}")


"""======================================Exectution======================================"""

person1 = SocialMedia() # object instantiation

# lets first try to log in first without signing up
person1.login("nikunj.dhaka", "kmsvcmdskv") # this will print to sign up first

# lets sign up
person1.signup("nikunj.dhaka", "kmsvcmdskv")

# lets log in again but with a wrong password:
person1.login("nikunj.dhaka", "wrong")

# lets log in with correct one
person1.login("nikunj.dhaka", "kmsvcmdskv")

# lets post some posts on feed:
person1.post("hey this is my first post")
person1.post("Good Morning Fellas")
person1.post("Happy mood")
person1.post("Feeling Good !!!!")

# lets look at the posts for person1 object:
person1.see_all_posts()

print("=============================")

person2 = SocialMedia()
person2.login("sonali singh", "kcwsmdlc")
person2.signup("sonali singh", "kcwsmdlc")
person2.login("sonali singh", "wrong")
person2.login("sonali singh", "kcwsmdlc")
person2.post("Yayaaya new profile")
person2.post("Night night")
person2.post("Goa Diaries")
person2.post("Feeling Awsm !!!!")
person2.see_all_posts()
```
