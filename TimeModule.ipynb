```python
"2025-10-05" # YYYY-DD-MM
"08:13 PM IST"
"9:47 AM CST"
"10:47 AM ET"
"UTC" # Universal Time Coordinated  -->  Jan 1, 1970

"2PM" # --> 12 hours format
"14" # --> 24 hours format
```

```python
import time
```

```python
print(time.time()) # will print UTC time
print(time.ctime()) # will print Current time
```

```python
print(time.tzname) # it will print(current timezone)
print("Standard timezone name:", time.tzname[0])
print("DST timezone name:", time.tzname[1])
```

```python
t1 = time.time()
time.sleep(2) # it will hold the interpretter for 2 second
t2 = time.time()
print(t2-t1)
```

```python
print(time.localtime())
```

```python
print(time.localtime().tm_year)
```

### time.strftime(format, t)


%M - minutes

%m - month

%d - date

%Y - YYYY Year format

%H - Hour

%S - Seconds

%y - YY year format


```python
print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
```
