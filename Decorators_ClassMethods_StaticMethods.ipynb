#Decorators
#### decorators are function that takes function as an argument
#### it changes the function by including some more lines

```python
import time

def time_it(func):
    def wrapper():
        t1 = time.time()
        print(f"Time start is : {t1}")
        func()
        t2 = time.time()
        print(f"Time start is : {t2}")
        t = t2-t1
        print(f"Time taken is : {t} seconds")
    return wrapper
```

```python
def research():
    print("Start")
    time.sleep(5)
    print("Complete")
```

```python
research = time_it(research)
```

```python
research()
```

## Python Sugar Syntax

##### A similar way to do the above is Sugar syntax

```python
@time_it
def research():
    print("Start")
    time.sleep(5)
    print("Complete")
```

```python
research()
```

```python
def log_time_it(func):
    def wrapper(*args, **kwargs): # *args dtype is tuple
        t1 = time.time()
        func(*args, **kwargs)
        t2 = time.time()
        t = t2-t1
        print(f"Time taken is : {t} seconds")
    return wrapper
```

```python
@log_time_it
def add(a,b,c,d):
    print(a+b+c)
    print(a+b+c)
    time.sleep(2)
    print(a+b+c)
    print(a+b+c)
```

```python
add(1,2,3,4)
```

#ClassMethod

```python
class B:
    a = 25 # class variable / Attribute
    def __init__(self):
        self.a = 100 # instance variable / Attibute
    def hello(self): # instance Methods
        print("hello world")
```

```python
# B.hello() - This will throw an error as "hello" is an instance method not a class method (it is bound to an object)
print(B.a) # this will print the class attribute, not the instance one
```

```python
b = B() # object of class B
```

```python
b.a # this will print the instance variable
```

```python
b.hello() # this will call the instance method
```

```python
class Memory:
    total_memory = 128

    @classmethod
    def update_memory(cls, m):
        cls.total_memory-=m

    def check_memory(self):
        print(Memory.total_memory)
```

```python
print(Memory.total_memory)
Memory.update_memory(100)
print(Memory.total_memory)
```

```python
m1 = Memory() # lets cate an object
m1.check_memory()
```

```python
m2 = Memory() # lets make another object
m2.check_memory()
```

```python
class DummyA:
    def __init__(self, name,age):
        self.name = name
        self.age = age

    @classmethod
    def from_fullname(cls, name): # "Monal 30"
        name, age = name.split()
        return cls(name, age)

    def print_data(self):
        print(self.name)
        print(self.age)
```

```python

obj_class_method = DummyA.from_fullname("Monal 30")
obj_class_method.print_data()
```

```python
class Person:
    population = 0  # class variable

    def __init__(self, name):
        self.name = name
        Person.population += 1

    @classmethod
    def get_population(cls):
        return cls.population

# Create some Person instances
p1 = Person("Alice")
p2 = Person("Bob")

# Call the class method
print(Person.get_population())  # Output: 2
```

# Static Methods

```python
# These methods does not depends on self or cls

class Calculator:

    @staticmethod
    def add(a, b):
        print(a+b)

    @staticmethod
    def multiply(a, b):
        print(a*b)

    def all(self, a, b):
        print(Calculator.add(a,b))
        print("doing something")
        print(Calculator.multiply(a,b))
```

```python
Calculator.add(1,2)
```

```python
class MathTools:
    @staticmethod
    def add(x, y):
        return x + y

# Call from class
print(MathTools.add(3, 5))  # Output: 8

# Call from instance
tool = MathTools()
print(tool.add(10, 20))     # Output: 30
```
